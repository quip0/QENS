## Error + Syndrome overlay (6.6.6 d=5)

```python
from qens import DepolarizingError, NoisySampler

code = ColorCode(distance=5, lattice_type="6.6.6")
result = NoisySampler(seed=42).sample_errors(code, DepolarizingError(p=0.10), shots=1)
error = result.sample_error(0)
syndrome = result.sample_syndrome(0)
print(f"Error:    {error}")
print(f"Syndrome: {syndrome}")

draw_lattice(code, syndrome=syndrome, error=error,
             title="6.6.6 d=5 — Error + Syndrome").show()
```

## Complex test: large codes with errors and syndromes

```python
from qens import ColorCode, SurfaceCode, DepolarizingError, NoisySampler, draw_lattice

sampler = NoisySampler(seed=99)
noise = DepolarizingError(p=0.15)

# --- 6.6.6 Color Code d=9 with heavy noise ---
cc_666 = ColorCode(distance=9, lattice_type="6.6.6")
r1 = sampler.sample_errors(cc_666, noise, shots=1)
print(f"6.6.6 d=9: {cc_666.num_data_qubits} data qubits, {len(cc_666._plaquettes)} plaquettes")
print(f"  Error:    {r1.sample_error(0)}")
print(f"  Syndrome: {r1.sample_syndrome(0)}")
draw_lattice(cc_666, syndrome=r1.sample_syndrome(0), error=r1.sample_error(0),
             title="6.6.6 Color Code (d=9) — p=0.15", show_labels=False).show()

# --- 4.8.8 Color Code d=7 with moderate noise ---
cc_488 = ColorCode(distance=7, lattice_type="4.8.8")
r2 = sampler.sample_errors(cc_488, DepolarizingError(p=0.10), shots=1)
print(f"\n4.8.8 d=7: {cc_488.num_data_qubits} data qubits, {len(cc_488._plaquettes)} plaquettes")
print(f"  Error:    {r2.sample_error(0)}")
print(f"  Syndrome: {r2.sample_syndrome(0)}")
draw_lattice(cc_488, syndrome=r2.sample_syndrome(0), error=r2.sample_error(0),
             title="4.8.8 Color Code (d=7) — p=0.10").show()

# --- Surface Code d=9 with heavy noise ---
sc = SurfaceCode(distance=9)
r3 = sampler.sample_errors(sc, noise, shots=1)
print(f"\nSurface d=9: {sc.num_data_qubits} data qubits, {sc.num_ancilla_qubits} ancillas")
print(f"  Error:    {r3.sample_error(0)}")
print(f"  Syndrome: {r3.sample_syndrome(0)}")
draw_lattice(sc, syndrome=r3.sample_syndrome(0), error=r3.sample_error(0),
             title="Surface Code (d=9) — p=0.15", show_labels=False).show()
```
